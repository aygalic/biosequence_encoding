```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
scores = pd.read_csv("../../k_silhouette.tsv", sep="\t", index_col=None)
```

```python
models = ["MLP AE", "VQ-VAE-1","VQ-VAE-2"]
filtering = ["No filtering", "File 1", "File 2"]
subgroup = ["All groups", "Healthy Control", "Idiopathic PD", "Genetic PD"]

scores["Model"] = np.repeat(models, 12)
scores["Filtering"] = np.repeat(filtering*3, 4)
scores["Subgroups"] = subgroup * 9
print(scores)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

sns.kdeplot(scores, x="silhouette_score", hue="optimal_k", ax=axes[0, 0])
sns.kdeplot(scores, x="silhouette_score", hue="Model", ax=axes[0, 1])
sns.kdeplot(scores, x="silhouette_score", hue="Filtering", ax=axes[1, 0])
sns.kdeplot(scores, x="silhouette_score", hue="Subgroups", ax=axes[1, 1])

# Adjust spacing between subplots (optional)
plt.tight_layout()

# Show the plots
plt.show()
```
